```python
import torch
import torch.nn as nn
import numpy as np
```

```python
data = [1,2,3]

data = torch.tensor(data)
print(data)
```

```python
data = np.array([1,1,1])
print(data)

data = torch.from_numpy(data)
print(data)
```

```python
print("Random matrix")

rand_matrix = torch.rand((3,3))
print(rand_matrix)
```

```python
test1 = torch.zeros((2,3))
print(test1)
```

```python
test2 = torch.tensor([[1,2,3],[4,5,6]])
test2 += torch.tensor([1])

print(test2)
```

```python
print("MSE sample")

a = torch.tensor([1.0, 2.0, 3.0])
b = torch.tensor([4.0, 5.0, 6.0])

loss = torch.mean((a - b) ** 2)
print(loss)

loss2 = nn.MSELoss()
print(loss2(a, b))

loss3 = nn.L1Loss()(a, b)
print(loss3)
```

```python
data = torch.randint(low = 0, high = 10, size = (3,4))

print(data)

arg1 = torch.argmax(data, dim=0)
print(arg1)

arg2 = torch.argmax(data, dim=1)
print(arg2)
```
