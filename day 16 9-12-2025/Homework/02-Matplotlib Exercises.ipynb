## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2

print(f"List x[:10]: {x[:10]}")
print(f"List y[:10]: {y[:10]}")
print(f"List z[:10]: {z[:10]}")
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Tạo figure với 2 axes ở vị trí khác nhau
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])  # Axes lớn
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # Axes nhỏ bên trong
```

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Tạo figure với 2 axes và vẽ (x,y) trên cả hai
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# Vẽ (x,y) trên cả 2 axes
ax1.plot(x, y)
ax2.plot(x, y)

fig
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
# Tạo figure với 2 axes, axes nhỏ lớn hơn
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Vẽ (x,y) trên cả 2 axes
ax1.plot(x, y)
ax2.plot(x, y)

fig
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Vẽ x,y,z với giới hạn xlim và ylim
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Axes lớn vẽ (x,z)
ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

# Axes nhỏ vẽ (x,y) với giới hạn
ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)

fig
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Tạo subplots với 1 hàng, 2 cột
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Vẽ (x,y) và (x,z) với linewidth và style khác nhau
fig, axes = plt.subplots(nrows=1, ncols=2)

# Vẽ (x,y) trên axes[0]
axes[0].plot(x, y, color='blue', linewidth=3, linestyle='-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# Vẽ (x,z) trên axes[1]
axes[1].plot(x, z, color='red', linewidth=2, linestyle='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# Thay đổi kích thước figure với figsize
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Vẽ (x,y) trên axes[0]
axes[0].plot(x, y, color='blue', linewidth=3, linestyle='-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# Vẽ (x,z) trên axes[1]
axes[1].plot(x, z, color='red', linewidth=2, linestyle='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
