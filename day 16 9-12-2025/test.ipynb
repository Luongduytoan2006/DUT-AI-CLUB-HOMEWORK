```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu Titanic
df = pd.read_csv('titanic.csv')

print("üìä D·ªÆ LI·ªÜU TITANIC")
print("="*60)
print(df.head(3))
print(f"\nK√≠ch th∆∞·ªõc: {df.shape[0]} h√†ng, {df.shape[1]} c·ªôt")
print(f"\nTh√¥ng tin c√°c c·ªôt:")
print(df.info())
```

```python
# L·ªçc d·ªØ li·ªáu ra
test_data = df[['Survived', 'Age']]
test_data = test_data.dropna()
print(test_data)
```

```python
# T·∫°o figure v·ªõi 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# B·∫£ng 1: Ng∆∞·ªùi s·ªëng s√≥t (Survived = 1)
survived_data = test_data[test_data['Survived'] == 1]
axes[0].hist(survived_data['Age'], bins=20, color='green', alpha=0.7, edgecolor='black')
axes[0].set_title('Ph√¢n b·ªë ƒë·ªô tu·ªïi - Ng∆∞·ªùi s·ªëng s√≥t', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Tu·ªïi', fontsize=12)
axes[0].set_ylabel('S·ªë l∆∞·ª£ng', fontsize=12)
axes[0].grid(True, alpha=0.3)

# B·∫£ng 2: Ng∆∞·ªùi ch·∫øt (Survived = 0)
died_data = test_data[test_data['Survived'] == 0]
axes[1].hist(died_data['Age'], bins=20, color='red', alpha=0.7, edgecolor='black')
axes[1].set_title('Ph√¢n b·ªë ƒë·ªô tu·ªïi - Ng∆∞·ªùi ch·∫øt', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Tu·ªïi', fontsize=12)
axes[1].set_ylabel('S·ªë l∆∞·ª£ng', fontsize=12)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"üìä Th·ªëng k√™:")
print(f"S·ªë ng∆∞·ªùi s·ªëng s√≥t: {len(survived_data)} (Tu·ªïi trung b√¨nh: {survived_data['Age'].mean():.1f})")
print(f"S·ªë ng∆∞·ªùi ch·∫øt: {len(died_data)} (Tu·ªïi trung b√¨nh: {died_data['Age'].mean():.1f})")
```
