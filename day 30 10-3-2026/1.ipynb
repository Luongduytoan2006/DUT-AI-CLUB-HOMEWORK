```python
import cv2
import matplotlib.pyplot as plt
import importlib.util

spec = importlib.util.spec_from_file_location("module1", "1.py")
module1 = importlib.util.module_from_spec(spec)
spec.loader.exec_module(module1)
ImageRestoration = module1.ImageRestoration

image_path = 'anh_cay_coi.jpg'
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

sobel_img = ImageRestoration.sobel(img)
laplacian_img = ImageRestoration.laplacian(img)
median_img = ImageRestoration.median_filter(img_rgb)
wiener_img = ImageRestoration.wiener_filter(img_rgb)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

axes[0].imshow(img_rgb)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(sobel_img, cmap='gray')
axes[1].set_title('Sobel')
axes[1].axis('off')

axes[2].imshow(laplacian_img, cmap='gray')
axes[2].set_title('Laplacian')
axes[2].axis('off')

axes[3].imshow(median_img)
axes[3].set_title('Median Filter')
axes[3].axis('off')

axes[4].imshow(wiener_img)
axes[4].set_title('Wiener Filter')
axes[4].axis('off')

fig.delaxes(axes[5])
plt.tight_layout()
plt.show()
```
