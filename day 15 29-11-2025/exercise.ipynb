- D√πng ƒë·ªÉ ki·ªÉm tra nhanh d·ªØ li·ªáu

- M·ªói l·∫ßn ch·∫°y s·∫Ω cho k·∫øt qu·∫£ kh√°c nhau

**Tham s·ªë:**
- `n`: S·ªë d√≤ng mu·ªën l·∫•y

```python
# L·∫•y 1 d√≤ng ng·∫´u nhi√™n (m·∫∑c ƒë·ªãnh)
print("üé≤ 1 d√≤ng ng·∫´u nhi√™n:")
print(df.sample())

print("\n" + "="*70 + "\n")

# L·∫•y 5 d√≤ng ng·∫´u nhi√™n
print("üé≤ 5 d√≤ng ng·∫´u nhi√™n:")
print(df.sample(5))

print("\n" + "="*70 + "\n")

# L·∫•y 10% d·ªØ li·ªáu ng·∫´u nhi√™n
print("üé≤ 10% d·ªØ li·ªáu ng·∫´u nhi√™n:")
print(df.sample(frac=0.1))
```

- S·ªë h√†ng v√† s·ªë c·ªôt
- T√™n c√°c c·ªôt
- Ki·ªÉu d·ªØ li·ªáu c·ªßa t·ª´ng c·ªôt (dtype)
- S·ªë gi√° tr·ªã non-null (kh√¥ng ph·∫£i NaN)
- Dung l∆∞·ª£ng b·ªô nh·ªõ

**D√πng ƒë·ªÉ:**

- ∆Ø·ªõc l∆∞·ª£ng dung l∆∞·ª£ng b·ªô nh·ªõ

## 5. L·∫•y t·∫≠p m·∫´u ng·∫´u nhi√™n

**M·ª•c ƒë√≠ch:** T·∫°o t·∫≠p test nh·ªè ƒë·ªÉ th·ª≠ nghi·ªám
- L·∫•y 10 d√≤ng ng·∫´u nhi√™n
- Gi·ªØ nguy√™n index g·ªëc ƒë·ªÉ truy v·∫øt

```python
# L·∫•y 10 d√≤ng ng·∫´u nhi√™n l√†m t·∫≠p test
print("L·∫•y 10 d√≤ng ng·∫´u nhi√™n l√†m t·∫≠p test:")
test_set = df.sample(10) 
print(test_set)

print(f"\n‚Üí Index g·ªëc: {test_set.index.tolist()}")
```

## 6. L∆∞u index g·ªëc v√† reset index

**V·∫•n ƒë·ªÅ:** Khi l·∫•y sample, index kh√¥ng li√™n t·ª•c (VD: 5, 23, 105, ...)

**Gi·∫£i ph√°p:**
1. **L∆∞u index g·ªëc v√†o c·ªôt m·ªõi**: `df.index` tr·∫£ v·ªÅ index hi·ªán t·∫°i
2. **Reset index**: `reset_index(drop=True)` t·∫°o index m·ªõi 0, 1, 2, ...

**L∆∞u √Ω v·ªÅ apply():**
- C√¢u l·ªánh `test_set["idx_root"] = test_set.index` **KH√îNG** d√πng apply()
- ƒê√¢y l√† g√°n tr·ª±c ti·∫øp (vectorized) - NHANH h∆°n apply()
- `test_set.index` tr·∫£ v·ªÅ Series c·ªßa index
- G√°n tr·ª±c ti·∫øp v√†o c·ªôt m·ªõi ‚Üí Kh√¥ng c·∫ßn v√≤ng l·∫∑p!

```python
test_set["idx_root"] = test_set.index

# drop=True: B·ªè c·ªôt index c≈©, kh√¥ng gi·ªØ l·∫°i
test_set = test_set.reset_index(drop=True)

print("\nSau khi reset index:")
print(test_set[["Name", "idx_root"]])
```

## 7. V√≠ d·ª• th·ª±c s·ª± d√πng apply() - Ph√¢n lo·∫°i h√†nh kh√°ch

**B√¢y gi·ªù m·ªõi l√† v√≠ d·ª• TH·ª∞C S·ª∞ d√πng apply()!**

M·ª•c ƒë√≠ch:
- Ph√¢n lo·∫°i d·ª±a tr√™n nhi·ªÅu ƒëi·ªÅu ki·ªán ph·ª©c t·∫°p
- C·∫ßn x·ª≠ l√Ω t·ª´ng h√†ng (row) ri√™ng bi·ªát d·ª±a tr√™n nhi·ªÅu tr∆∞·ªùng d·ªØ li·ªáu
- D√πng `apply(axis=1)` ƒë·ªÉ √°p d·ª•ng h√†m tr√™n t·ª´ng h√†ng


```python
def classify_passenger(row):
    age = row['Age']
    pclass = row['Pclass']
    fare = row['Fare']
    sex = row['Sex']
    
    if pd.isna(age):
        return "Unknown"
    elif age < 12:
        return "Child"
    elif age < 18:
        if pclass == 1:
            return "Rich Teen"
        else:
            return "Teen"
    elif sex == 'female' and pclass == 1 and fare > 50:
        return "VIP Lady"
    elif pclass == 1:
        return "First Class Adult"
    elif age > 60:
        return "Senior"
    else:
        return "Regular Adult"

# √ÅP D·ª§NG h√†m cho T·ª™NG H√ÄNG v·ªõi axis=1
test_set['PassengerType'] = test_set.apply(classify_passenger, axis=1)

print("üè∑Ô∏è  Ph√¢n lo·∫°i h√†nh kh√°ch b·∫±ng apply():")
print("="*70)
print(test_set[['Name', 'Age', 'Sex', 'Pclass', 'Fare', 'PassengerType']])
```

- mean: Trung b√¨nh
- std: ƒê·ªô l·ªách chu·∫©n
- min, max: Gi√° tr·ªã nh·ªè nh·∫•t, l·ªõn nh·∫•t
- 25%, 50%, 75%: Ph√¢n v·ªã (quartiles)

```python
# Th·ªëng k√™ m√¥ t·∫£
print("üìä Th·ªëng k√™ m√¥ t·∫£ c√°c c·ªôt s·ªë:")
print("="*70)
print(df.describe())

print("\nüí° Gi·∫£i th√≠ch:")
print("- count: S·ªë gi√° tr·ªã kh√¥ng ph·∫£i NaN (Age c√≥ 714/891 ‚Üí thi·∫øu 177 gi√° tr·ªã)")
print("- mean: Tu·ªïi trung b√¨nh ~29.7, gi√° v√© trung b√¨nh ~32.2")
print("- std: ƒê·ªô l·ªách chu·∫©n ‚Üí Fare c√≥ std=49.7 (bi·∫øn ƒë·ªông l·ªõn!)")
print("- 50%: Trung v·ªã (median) ‚Üí Tu·ªïi gi·ªØa l√† 28")
print("- 75% - 25% = IQR (Interquartile Range) ‚Üí Ph√°t hi·ªán outliers")
```

## 10. X·ª≠ l√≠ Nan

**ƒê·∫øm gi√° tr·ªã Nan trong m·ªói c·ªôt:**
- `df.isna().sum()`: Tr·∫£ v·ªÅ s·ªë l∆∞·ª£ng Nan trong m·ªói c·ªôt

**Lo·∫°i b·ªè h√†ng c√≥ Nan:**
- `df.dropna()`: Lo·∫°i b·ªè c√°c h√†ng c√≥ √≠t nh·∫•t m·ªôt gi√° tr·ªã NaN

**Thay th·∫ø Nan b·∫±ng gi√° tr·ªã kh√°c:**
- `df.fillna(value)`: Thay th·∫ø t·∫•t c·∫£ Nan b·∫±ng `value`
- N√™n xem x√©t thay b·∫±ng c√°c gi√° tr·ªã sau: 0, -1, mean, median, mode, "Unknown"

```python
print(f"S·ªë l∆∞∆°ng Nan trong m·ªói c·ªôt l√†:\n {df.isna().sum()}")
```

```python
print(f"K√≠ch th∆∞·ªõc df ban ƒë·∫ßu:", df.shape)

df_clean = df.dropna()   # Lo·∫°i b·ªè c√°c h√†ng c√≥ NaN
print(f"K√≠ch th∆∞·ªõc df sau khi lo·∫°i b·ªè NaN:", df_clean.shape)

print("=" * 72)
print("D·ªØ li·ªáu sau khi lo·∫°i b·ªè NaN:\n", df_clean)
```

```python
# L·∫•y ra 100 d·ªØ li·ªáu test t·ª´ df ƒë·ªÉ th·ª≠ thay th·∫ø 
df_test_nan = df.sample(100, random_state=42)
print("D·ªØ li·ªáu test ban ƒë·∫ßu v·ªõi NaN:\n", df_test_nan.sample(3))

print("=" * 72)
print(f"S·ªë l∆∞∆°ng Nan trong m·ªói c·ªôt l√†:\n {df_test_nan.isna().sum()}")


df_test_nan['Age'] = df_test_nan['Age'].fillna(df_clean['Age'].mean())      # ƒëi·ªÅn b·∫±ng trung b√¨nh c·ªßa d·ªØ li·ªáu s·∫°ch
df_test_nan['Age'] = df_test_nan['Age'].fillna(df_clean['Age'].median())    # ƒëi·ªÅn b·∫±ng trung v·ªã c·ªßa d·ªØ li·ªáu s·∫°ch
df_test_nan['Cabin'] = df_test_nan['Cabin'].fillna('Unknown')  # ƒëi·ªÅn b·∫±ng 'Unknown'


print(f"S·ªë l∆∞∆°ng Nan trong m·ªói c·ªôt sau khi fill l√†:\n {df_test_nan.isna().sum()}")
```

## 11. X·ª≠ l√≠ Tr√πng l·∫∑p

**Ki·ªÉm tra tr√πng l·∫∑p**
- `df.duplicated()`: Tr·∫£ v·ªÅ Series boolean, True n·∫øu h√†ng tr√πng l·∫∑p v·ªõi h√†ng tr∆∞·ªõc ƒë√≥
- `df.duplicated().sum()`: ƒê·∫øm s·ªë l∆∞·ª£ng b·∫£n ghi tr√πng l·∫∑p

**Lo·∫°i b·ªè tr√πng l·∫∑p**
- `df.drop_duplicates(inplace = True)`: Lo·∫°i b·ªè c√°c h√†ng tr√πng l·∫∑p, gi·ªØ l·∫°i h√†ng ƒë·∫ßu ti√™n t·ª´ tr√™n xu·ªëng
L∆∞u √Ω: Tham s·ªë (inplace = True) s·∫Ω ƒë·∫£m b·∫£o r·∫±ng ph∆∞∆°ng th·ª©c kh√¥ng t·∫°o ra m·ªôt DataFrame m·ªõi, m√† s·∫Ω x√≥a c√°c d√≤ng tr√πng l·∫∑p tr·ª±c ti·∫øp tr√™n DataFrame g·ªëc.

```python
# Tr√πng l·∫∑p
print(df.duplicated())
print(f"S·ªë l∆∞·ª£ng b·∫£n ghi tr√πng l·∫∑p: {df.duplicated().sum()}")
```

```python
# L·∫•y 5 d√≤ng ƒë·∫ßu v√† gh√©p th√™m 1 d√≤ng ƒë·∫ßu
df_test_dup = pd.concat([df.head(5), df.head(1)], ignore_index=False)
print(df_test_dup['Name'])

print("\n" + "="*70 + "\n")

# D√πng iloc v√† list index
df_test_dup2 = df.iloc[[0, 1, 2, 3, 4, 0]]  # index 0 xu·∫•t hi·ªán 2 l·∫ßn
print(df_test_dup2['Name'])

print("\n" + "="*70 + "\n")

print(df_test_dup.duplicated())
print(f"S·ªë l∆∞·ª£ng b·∫£n ghi tr√πng l·∫∑p trong df_test_dup: {df_test_dup.duplicated().sum()}")

print("\n" + "="*70 + "\n")

df_test_dup = df_test_dup.drop_duplicates()
print("D·ªØ li·ªáu sau khi lo·∫°i b·ªè b·∫£n tr√πng l·∫∑p:\n", df_test_dup['Name'])
```
