## 5. L·∫•y t·∫≠p m·∫´u ng·∫´u nhi√™n

**M·ª•c ƒë√≠ch:** T·∫°o t·∫≠p test nh·ªè ƒë·ªÉ th·ª≠ nghi·ªám

## 6. L∆∞u index g·ªëc v√† reset index

**V·∫•n ƒë·ªÅ:** Khi l·∫•y sample, index kh√¥ng li√™n t·ª•c (VD: 5, 23, 105, ...)

**Gi·∫£i ph√°p:**
1. **L∆∞u index g·ªëc v√†o c·ªôt m·ªõi**: `df.index` tr·∫£ v·ªÅ index hi·ªán t·∫°i
2. **Reset index**: `reset_index(drop=True)` t·∫°o index m·ªõi 0, 1, 2, ...

**L∆∞u √Ω v·ªÅ apply():**
- C√¢u l·ªánh `test_set["idx_root"] = test_set.index` **KH√îNG** d√πng apply()
- ƒê√¢y l√† g√°n tr·ª±c ti·∫øp (vectorized) - NHANH h∆°n apply()
- `test_set.index` tr·∫£ v·ªÅ Series c·ªßa index
- G√°n tr·ª±c ti·∫øp v√†o c·ªôt m·ªõi ‚Üí Kh√¥ng c·∫ßn v√≤ng l·∫∑p!

````python
test_set["idx_root"] = test_set.index

# drop=True: B·ªè c·ªôt index c≈©, kh√¥ng gi·ªØ l·∫°i
test_set = test_set.reset_index(drop=True)

print("\nSau khi reset index:")
print(test_set[["Name", "idx_root"]])
````

## 7. V√≠ d·ª• th·ª±c s·ª± d√πng apply() - Ph√¢n lo·∫°i h√†nh kh√°ch

**B√¢y gi·ªù m·ªõi l√† v√≠ d·ª• TH·ª∞C S·ª∞ d√πng apply()!**

M·ª•c ƒë√≠ch:
- Ph√¢n lo·∫°i d·ª±a tr√™n nhi·ªÅu ƒëi·ªÅu ki·ªán ph·ª©c t·∫°p
- C·∫ßn x·ª≠ l√Ω t·ª´ng h√†ng (row) ri√™ng bi·ªát d·ª±a tr√™n nhi·ªÅu tr∆∞·ªùng d·ªØ li·ªáu
- D√πng `apply(axis=1)` ƒë·ªÉ √°p d·ª•ng h√†m tr√™n t·ª´ng h√†ng


````python
def classify_passenger(row):
    age = row['Age']
    pclass = row['Pclass']
    fare = row['Fare']
    sex = row['Sex']
    
    if pd.isna(age):
        return "Unknown"
    elif age < 12:
        return "Child"
    elif age < 18:
        if pclass == 1:
            return "Rich Teen"
        else:
            return "Teen"
    elif sex == 'female' and pclass == 1 and fare > 50:
        return "VIP Lady"
    elif pclass == 1:
        return "First Class Adult"
    elif age > 60:
        return "Senior"
    else:
        return "Regular Adult"


test_set['PassengerType'] = test_set.apply(classify_passenger, axis=1)

print("Ph√¢n lo·∫°i h√†nh kh√°ch b·∫±ng apply():")
print(test_set[['Name', 'Age', 'Sex', 'Pclass', 'Fare', 'PassengerType']])
````

## 12. S·ª≠ d·ª•ng groupby, SQL

GROUPBY CH·ªà S·ª¨ D·ª§NG KHI D·ªÆ LI·ªÜU, PH∆Ø∆†NG TH·ª®C X·ª¨ L√ç C·∫¶N THEO NH√ìM, PH·ª®C T·∫†P  

**C√°c ph∆∞∆°ng th·ª©c c∆° b·∫£n**
- `df.groupby('column')`: Nh√≥m d·ªØ li·ªáu theo gi√° tr·ªã trong c·ªôt
- `df.groupby('column').agg({'col1': 'mean', 'col2': 'sum'})`: T√≠nh to√°n c√°c th·ªëng k√™ t·ªïng h·ª£p cho t·ª´ng nh√≥m
- `df.groupby('Embarked').agg({'Age': 'mean', 'Fare': 'sum'})`: T√≠nh tu·ªïi trung b√¨nh v√† t·ªïng ti·ªÅn v√© theo c·∫£ng l√™n t√†u


**D√πng groupby ƒë·ªÉ nh√≥m v√† t√≠nh theo nhu c·∫ßu, ph∆∞∆°ng tr√¨nh**
- `df.groupby('Pclass')['Age'].mean()`: T√≠nh tu·ªïi trung b√¨nh theo t·ª´ng h·∫°ng v√©
- `df.groupby('Survived')['Fare'].sum()`: T√≠nh t·ªïng ti·ªÅn v√© theo tr·∫°ng th√°i s·ªëng s√≥t
- `df.groupby(['Pclass', 'Sex'])['Survived'].mean()`: T√≠nh t·ª∑ l·ªá s·ªëng s√≥t theo h·∫°ng v√© v√† gi·ªõi t√≠nh
- `df.groupby('Pclass').function()`: √Åp d·ª•ng h√†m t√πy ch·ªânh cho t·ª´ng nh√≥m

**D√πng SQL v·ªõi pandas**
- S·ª≠ d·ª•ng th∆∞ vi·ªán `pandasql` ƒë·ªÉ ch·∫°y c√¢u l·ªánh SQL tr√™n DataFrame

````python
# S·ª≠ d·ª•ng b·ªô test_set ƒë·ªÉ l√†m

print("S·ª≠ d·ª•ng b·ªô test_set ƒë·ªÉ l√†m")
print(test_set)
````

````python
# Group v√† ƒë·∫øm xem m·ªói gi·ªõi t√≠nh bao nhi√™u ng∆∞·ªùi
print("Group v√† ƒë·∫øm xem m·ªói gi·ªõi t√≠nh bao nhi√™u ng∆∞·ªùi")
print(test_set.groupby('Sex').size())

print("\n" + "="*70 + "\n")

# Group tu·ªïi ƒë·ªÉ xem c√≥ bao nhi√™u ng∆∞·ªùi trong m·ªói nh√≥m tu·ªïi
print("Group tu·ªïi ƒë·ªÉ xem c√≥ bao nhi√™u ng∆∞·ªùi trong m·ªói nh√≥m tu·ªïi")

def age_group(age):
    if pd.isna(age):
        return "Unknown"
    elif age < 12:
        return "Child"
    elif age < 18:
        return "Teen"
    elif age < 35:
        return "Young Adult"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

print(test_set.groupby(test_set['Age'].apply(age_group)).size())
````

````python
# Group nhi·ªÅu nh√≥m - ƒê·∫øm s·ªë ng∆∞·ªùi theo gi·ªõi t√≠nh v√† h·∫°ng v√©
print("Group nhi·ªÅu nh√≥m - ƒê·∫øm s·ªë ng∆∞·ªùi:")
print(test_set.groupby(['Sex', 'Pclass']).size())

print("\n" + "="*70 + "\n")

# T√≠nh tu·ªïi trung b√¨nh theo t·ª´ng nh√≥m (Sex, Pclass)
print("Tu·ªïi trung b√¨nh theo gi·ªõi t√≠nh v√† h·∫°ng v√©:")
print(test_set.groupby(['Sex', 'Pclass'])['Age'].mean())

print("\n" + "="*70 + "\n")

# C√ÅCH 1: D√πng apply v·ªõi groupby ƒë·ªÉ ph√¢n lo·∫°i (SAI - kh√¥ng n√™n d√πng)
# print(test_set.groupby(['Sex', 'Pclass']).apply(classify_passenger))  # ‚Üê SAI

# C√ÅCH 2: D√πng apply TR·ª∞C TI·∫æP tr√™n DataFrame (ƒê√öNG)
def classify_passenger(row):
    sex = row['Sex']
    pclass = row['Pclass']
    if sex == 'female':
        if pclass == 1:
            return "Ph√∫ b√†"
        elif pclass == 2:
            return "Qu√Ω c√¥"
        else:
            return "Th∆∞·ªùng d√¢n (n·ªØ)"
    else:
        if pclass == 1:
            return "Qu√Ω √¥ng"
        elif pclass == 2:
            return "C√¥ng t·ª≠"
        else:
            return "Th∆∞·ªùng d√¢n (nam)"

# √Åp d·ª•ng h√†m cho T·ª™NG H√ÄNG (axis=1)
test_set['Category'] = test_set.apply(classify_passenger, axis=1)

print("Ph√¢n lo·∫°i h√†nh kh√°ch theo gi·ªõi t√≠nh v√† h·∫°ng v√©:")
print(test_set[['Name', 'Sex', 'Pclass', 'Category']])

print("\n" + "="*70 + "\n")

# Sau khi c√≥ c·ªôt Category, c√≥ th·ªÉ groupby ƒë·ªÉ ƒë·∫øm
print("ƒê·∫øm s·ªë ng∆∞·ªùi m·ªói lo·∫°i:")
print(test_set.groupby('Category').size())
````

````python
# In ra nh·ªØng ng∆∞·ªùi c√≤n s·ªëng tr√™n 30 tu·ªïi v√† l√† nam
print("In ra nh·ªØng ng∆∞·ªùi c√≤n s·ªëng tr√™n 30 tu·ªïi v√† l√† nam")
survived_men_over_30 = df[(df['Survived'] == 1) & (df['Sex'] == 'male') & (df['Age'] > 30)]
print(survived_men_over_30)
````

## üìñ S·ª∞ KH√ÅC BI·ªÜT GI·ªÆA `groupby()` V√Ä `apply()`

### üéØ T√ìM T·∫ÆT NHANH

| Ti√™u ch√≠ | `groupby()` | `apply()` |
