```python
import pandas as pd
import numpy as np
df = pd.read_csv('IRIS.csv')
print(df)
```

```python
y = df['species']
print(y)

x = df.drop(columns=['species'])
print(x)
```

```python
x = x.to_numpy()
print(x)
y = y.to_numpy()
print(y)

print( x.shape,  y.shape)
```

```python
# T√°ch d·ªØ li·ªáu th√†nh t·∫≠p hu·∫•n luy·ªán v√† t·∫≠p ki·ªÉm tra
x_train = x[:120]
y_train = y[:120]
x_test = x[120:]
y_test = y[120:]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python

def calculate_distance(x_new, x_train):
    distances = np.sqrt(np.sum((x_new - x_train) ** 2, axis=1))
    return distances
```

```python
# Test th·ª≠
x_new = np.array([5.0, 3.0, 1.5, 0.3])
distances = calculate_distance(x_new, x_train)
print(f"Kho·∫£ng c√°ch t·ª´ ƒëi·ªÉm m·ªõi ƒë·∫øn {len(distances)} ƒëi·ªÉm train:")
print(distances[:10])
```

```python
k = 5

# np.argsort(distances) tr·∫£ v·ªÅ INDEX c·ªßa c√°c ph·∫ßn t·ª≠ khi s·∫Øp x·∫øp tƒÉng d·∫ßn
# [:k] l·∫•y k ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n (k ƒëi·ªÉm g·∫ßn nh·∫•t)
top_k_indices = np.argsort(distances)[:k]

print(f"Ch·ªçn k = {k} h√†ng x√≥m g·∫ßn nh·∫•t")
print(f"Index c·ªßa {k} ƒëi·ªÉm g·∫ßn nh·∫•t: {top_k_indices}")
print(f"Kho·∫£ng c√°ch c·ªßa {k} ƒëi·ªÉm ƒë√≥: {distances[top_k_indices]}")
```

```python
# L·∫•y label c·ªßa k ƒëi·ªÉm g·∫ßn nh·∫•t
labels = y_train[top_k_indices]
print(f"Nh√£n c·ªßa {k} h√†ng x√≥m g·∫ßn nh·∫•t: {labels}")
```

```python
# B·ªè phi·∫øu
dataset_labels = np.unique(y_train)
print(f"C√°c lo·∫°i hoa trong dataset: {dataset_labels}")

counts = {}
for label in dataset_labels:
    counts[label] = 0

for label in labels:
    counts[label] += 1

print(f"\nK·∫øt qu·∫£ b·ªè phi·∫øu: {counts}")

# Ch·ªçn nh√£n c√≥ s·ªë phi·∫øu cao nh·∫•t
predicted_label = max(counts, key=counts.get)
print(f"‚Üí D·ª± ƒëo√°n: ƒëi·ªÉm m·ªõi thu·ªôc lo·∫°i '{predicted_label}'")
```

```python

# Pipeline KNN
def pipeline_knn(x_new, x_train, y_train, k):
    # B∆∞·ªõc 1: T√≠nh kho·∫£ng c√°ch
    distances = calculate_distance(x_new, x_train)
    
    # B∆∞·ªõc 2: T√¨m k ƒëi·ªÉm g·∫ßn nh·∫•t
    top_k_indices = np.argsort(distances)[:k]
    
    # B∆∞·ªõc 3: L·∫•y nh√£n c·ªßa k ƒëi·ªÉm ƒë√≥
    labels = y_train[top_k_indices]
    
    # B∆∞·ªõc 4: ƒê·∫øm v√† ch·ªçn nh√£n xu·∫•t hi·ªán nhi·ªÅu nh·∫•t
    dataset_labels = np.unique(y_train)
    counts = {}
    for label in dataset_labels:
        counts[label] = 0
    
    for label in labels:
        counts[label] += 1
    
    predicted_label = max(counts, key=counts.get)
    return predicted_label

x_new = np.array([5.0, 3.0, 1.5, 0.3])
prediction = pipeline_knn(x_new, x_train, y_train, k=3)
print(f"D·ª± ƒëo√°n cho ƒëi·ªÉm {x_new}: {prediction}")
```

```python

y_pred_test = []
k = 3

print(f"ƒêang d·ª± ƒëo√°n cho {len(x_test)} ƒëi·ªÉm test v·ªõi k={k}...")

for i, x_test_point in enumerate(x_test):
    prediction = pipeline_knn(x_test_point, x_train, y_train, k)
    y_pred_test.append(prediction)
    
    # In ti·∫øn tr√¨nh
    if (i + 1) % 10 == 0:
        print(f"ƒê√£ d·ª± ƒëo√°n {i + 1}/{len(x_test)} ƒëi·ªÉm")

y_pred_test = np.array(y_pred_test)
print(f"\n‚úì Ho√†n th√†nh! ƒê√£ d·ª± ƒëo√°n xong {len(y_pred_test)} ƒëi·ªÉm")
```

```python
# T√≠nh ƒë·ªô ch√≠nh x√°c (accuracy)
correct = np.sum(y_pred_test == y_test)  # ƒê·∫øm s·ªë d·ª± ƒëo√°n ƒë√∫ng
accuracy = correct / len(y_test) * 100

print(f"S·ªë d·ª± ƒëo√°n ƒë√∫ng: {correct}/{len(y_test)}")
print(f"ƒê·ªô ch√≠nh x√°c (Accuracy): {accuracy:.2f}%")
```

```python
print("Th·ª≠ nghi·ªám v·ªõi nhi·ªÅu gi√° tr·ªã k kh√°c nhau:\n")

best_k = 0
best_accuracy = 0

for k in [1, 3, 5, 7, 9, 11, 15, 21]:
    # D·ª± ƒëo√°n v·ªõi k n√†y
    y_pred = []
    for x_test_point in x_test:
        prediction = pipeline_knn(x_test_point, x_train, y_train, k)
        y_pred.append(prediction)
    
    y_pred = np.array(y_pred)
    
    # T√≠nh accuracy
    correct = np.sum(y_pred == y_test)
    accuracy = correct / len(y_test) * 100
    
    print(f"k={k:2d}: Accuracy = {accuracy:.2f}%")
    
    # L∆∞u k t·ªët nh·∫•t
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f"\nüèÜ Gi√° tr·ªã k t·ªët nh·∫•t: k={best_k} v·ªõi accuracy={best_accuracy:.2f}%")
```

```python
# ============================================
# T√ìM T·∫ÆT THU·∫¨T TO√ÅN KNN
# ============================================
#
# KNN (K-Nearest Neighbors) - K H√†ng X√≥m G·∫ßn Nh·∫•t
#
# √ù t∆∞·ªüng: "Cho t√¥i bi·∫øt b·∫°n c·ªßa b·∫°n l√† ai, t√¥i s·∫Ω n√≥i b·∫°n l√† ng∆∞·ªùi nh∆∞ th·∫ø n√†o"
#
# C√°c b∆∞·ªõc:
# 1. T√≠nh kho·∫£ng c√°ch t·ª´ ƒëi·ªÉm c·∫ßn d·ª± ƒëo√°n ƒë·∫øn T·∫§T C·∫¢ ƒëi·ªÉm trong training set
# 2. S·∫Øp x·∫øp v√† ch·ªçn k ƒëi·ªÉm g·∫ßn nh·∫•t
# 3. L·∫•y nh√£n c·ªßa k ƒëi·ªÉm ƒë√≥
# 4. B·ªè phi·∫øu: ch·ªçn nh√£n xu·∫•t hi·ªán nhi·ªÅu nh·∫•t
#
# V√≠ d·ª• v·ªõi k=3:
# - ƒêi·ªÉm m·ªõi c√≥ 3 h√†ng x√≥m g·∫ßn nh·∫•t: [setosa, setosa, versicolor]
# - B·ªè phi·∫øu: setosa=2 phi·∫øu, versicolor=1 phi·∫øu
# - ‚Üí D·ª± ƒëo√°n: setosa (v√¨ c√≥ 2/3 phi·∫øu)
#
# L∆∞u √Ω:
# - k nh·ªè: nh·∫°y c·∫£m v·ªõi nhi·ªÖu, d·ªÖ overfitting
# - k l·ªõn: m∆∞·ª£t h∆°n nh∆∞ng c√≥ th·ªÉ b·ªã underfitting
# - k th∆∞·ªùng ch·ªçn l·∫ª ƒë·ªÉ tr√°nh h√≤a phi·∫øu
#
print("‚úì KNN ho√†n ch·ªânh! B·∫°n ƒë√£ hi·ªÉu c√°ch KNN ho·∫°t ƒë·ªông r·ªìi ƒë·∫•y!")
```
